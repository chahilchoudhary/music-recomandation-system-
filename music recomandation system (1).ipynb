```python
data1['lyrics'] = data1['lyrics'].str.lower().replace(r'[^a-zA-Z0-9]','').replace(r'\n','', regex=True)
```

# Doing tokenization

```python
import nltk
nltk.download('punkt')
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

stemming = []
def tokenization(txt):
    tokens = nltk.word_tokenize(txt)
    for w in tokens:
        stemming.append(ps.stem(w))
        
    return " ".join(stemming)
```

```python
tokenization('this is my word loving loved')
```

```python
data1['lyrics'] = data1['lyrics'].apply(lambda x: tokenization(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Tfid = TfidfVectorizer(stop_words='english')
matrix = Tfid.fit_transform(data1['lyrics'])
```

```python
matrix.shape
```

```python
similarity = cosine_similarity(matrix)
```

```python
similarity[0]
```

```python
data1[data1['track_name']=='tumbling tumbleweeds']
```

```python
sorted(list(enumerate(similarity[0])),reverse=False, key=lambda x:x[1])
```

```python
def recommendation(track_name):
    data1[data1['track_name']==track_name].index[0]
    distances=sorted(list(enumerate(similarity[0])),reverse=False, key=lambda x:x[1])
    
    
    songs = []
    for i in distances[1:21]:
        songs.append(data1.iloc[i[0]].track_name)
        
        
    return songs
```
